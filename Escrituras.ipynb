```python
import altair as alt
import geojson
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/BJRomPal/escrituras/main/barrios.geojson"
url2 = "https://raw.githubusercontent.com/BJRomPal/escrituras/main/centros_barrios.csv"
```

```python
centroides = pd.read_csv(url2, encoding="latin-1")
centroides
```

```python
centroides.replace("NUÃEZ", "NUÑEZ", inplace = True)
centroides
```

```python
data_geojson_remote = alt.Data(url=url, format=alt.DataFormat(property='features',type='json'))

background = alt.Chart(data_geojson_remote).mark_geoshape(
    fill='lightgray',
    stroke='white'
).properties(
    width=500,
    height=500
).project('equirectangular')

points = alt.Chart(centroides).mark_circle().encode(
    longitude='longitud:Q',
    latitude='latitud:Q',
    size=alt.Size('Cantidad:Q', title='Cantidad de Escrituras'),
    color=alt.value('steelblue'),
    tooltip=['Barrio:N','Cantidad:Q']
).properties(
    title='Escrituras Septiembre 2020'
)

background + points
```

```python
chart = background + points
chart.save('escrituras.html')
```
